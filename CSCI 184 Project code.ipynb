```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Embedding, Flatten, Concatenate
from tensorflow.keras.models import Model

# Defining input layers for each feature
input_track_id = Input(shape=(1,), name='track_id')
input_danceability = Input(shape=(1,), name='danceability')
input_tempo = Input(shape=(1,), name='tempo')
input_acousticness = Input(shape=(1,), name='acousticness')
input_instrumentalness = Input(shape=(1,), name='instrumentalness')
input_speechiness = Input(shape=(1,), name='speechiness')
input_liveness = Input(shape=(1,), name='liveness')
input_energy = Input(shape=(1,), name='energy')
input_genre = Input(shape=(1,), name='genre')
input_genre_top = Input(shape=(1,), name='genre_top')
input_genres_all = Input(shape=(1,), name='genres_all')

# Defining the embedding size for categorical features
embedding_size = 8

# Embedding layers for categorical features
embedding_genre = Embedding(input_dim=584, output_dim=embedding_size)(input_genre)
embedding_genre_top = Embedding(input_dim=13, output_dim=embedding_size)(input_genre_top)
embedding_genres_all = Embedding(input_dim=541, output_dim=embedding_size)(input_genres_all)

# Flattening the embedding layers
flatten_genre = Flatten()(embedding_genre)
flatten_genre_top = Flatten()(embedding_genre_top)
flatten_genres_all = Flatten()(embedding_genres_all)

# Concatenating all input layers
concatenated_inputs = Concatenate()([input_track_id, input_danceability, input_tempo, input_acousticness,
                                    input_instrumentalness, input_speechiness, input_liveness, input_energy,
                                    flatten_genre, flatten_genre_top, flatten_genres_all])

# Hidden layers
hidden1 = Dense(128, activation='relu')(concatenated_inputs)
hidden2 = Dense(64, activation='relu')(hidden1)
hidden3 = Dense(32, activation='relu')(hidden2)

# Output layer
output = Dense(13, activation='softmax')(hidden3)  # Replace `num_classes` with the number of genre classes

# Creating the model
model = Model(inputs=[input_track_id, input_danceability, input_tempo, input_acousticness,
                      input_instrumentalness, input_speechiness, input_liveness, input_energy,
                      input_genre, input_genre_top, input_genres_all],
              outputs=output)

# Compiling the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Printing the summary of the model
model.summary()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical

# Converting input features to NumPy arrays and convert to float32
track_id_train = np.array(cleaned_df['track_id'], dtype=np.float32)
danceability_train = np.array(cleaned_df['danceability'], dtype=np.float32)
tempo_train = np.array(cleaned_df['tempo'], dtype=np.float32)
acousticness_train = np.array(cleaned_df['acousticness'], dtype=np.float32)
instrumentalness_train = np.array(cleaned_df['instrumentalness'], dtype=np.float32)
speechiness_train = np.array(cleaned_df['speechiness'], dtype=np.float32)
liveness_train = np.array(cleaned_df['liveness'], dtype=np.float32)
energy_train = np.array(cleaned_df['energy'], dtype=np.float32)

# Converting target variable to NumPy array
genre_top_train = np.array(cleaned_df['genre_top'])

# Encoding the target variable
label_encoder = LabelEncoder()
genre_top_train_encoded = label_encoder.fit_transform(genre_top_train)
num_classes = len(label_encoder.classes_)
y_train_encoded = to_categorical(genre_top_train_encoded)

# Combining input features into a single array
X_train = np.vstack((track_id_train, danceability_train, tempo_train, acousticness_train,
                     instrumentalness_train, speechiness_train, liveness_train, energy_train)).T

# Splitting the data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_train,
    y_train_encoded,
    test_size=0.2,
    random_state=42
)

# Defining input features
inputs = []
for i in range(X_train.shape[1]):
    input_feature = Input(shape=(1,))
    inputs.append(input_feature)

# Concatenating input features
concatenated_inputs = Concatenate()(inputs)

# Defining hidden layers
hidden = Dense(256, activation='relu')(concatenated_inputs)

# Defining output layer
output = Dense(num_classes, activation='softmax')(hidden)

# Creating the model
model = Model(inputs=inputs, outputs=output)


# Compiling the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


model.fit(
    [X_train[:, i] for i in range(X_train.shape[1])],
    y_train,
    epochs=10,
    batch_size=32,
    validation_data=([X_val[:, i] for i in range(X_val.shape[1])], y_val),
)
```

```python
from tensorflow.keras.callbacks import LearningRateScheduler
from tensorflow.keras.models import load_model


track_id_train = np.array(cleaned_df['track_id'], dtype=np.float32)
danceability_train = np.array(cleaned_df['danceability'], dtype=np.float32)
tempo_train = np.array(cleaned_df['tempo'], dtype=np.float32)
acousticness_train = np.array(cleaned_df['acousticness'], dtype=np.float32)
instrumentalness_train = np.array(cleaned_df['instrumentalness'], dtype=np.float32)
speechiness_train = np.array(cleaned_df['speechiness'], dtype=np.float32)
liveness_train = np.array(cleaned_df['liveness'], dtype=np.float32)
energy_train = np.array(cleaned_df['energy'], dtype=np.float32)


genre_top_train = np.array(cleaned_df['genre_top'])


label_encoder = LabelEncoder()
genre_top_train_encoded = label_encoder.fit_transform(genre_top_train)
num_classes = len(label_encoder.classes_)
y_train_encoded = to_categorical(genre_top_train_encoded)


X_train = np.vstack((track_id_train, danceability_train, tempo_train, acousticness_train,
                     instrumentalness_train, speechiness_train, liveness_train, energy_train)).T


X_train, X_val, y_train, y_val = train_test_split(
    X_train,
    y_train_encoded,
    test_size=0.2,
    random_state=42
)


inputs = []
for i in range(X_train.shape[1]):
    input_feature = Input(shape=(1,))
    inputs.append(input_feature)


concatenated_inputs = Concatenate()(inputs)

# Defining new hidden layer and output layer
hidden = Dense(100000, activation='relu')(concatenated_inputs)
output = Dense(num_classes, activation='softmax')(hidden)


model = Model(inputs=inputs, outputs=output)

# Defining a function to schedule the learning rate
def lr_schedule(epoch):
    initial_learning_rate = 0.005  # Initial learning rate
    decay_rate = 0.1  # Learning rate decay
    decay_steps = 5  # Number of epochs after which to decay the learning rate
    lr = initial_learning_rate * (decay_rate ** (epoch // decay_steps))
    return lr


lr_scheduler = LearningRateScheduler(lr_schedule)


model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


model.fit(
    [X_train[:, i] for i in range(X_train.shape[1])],
    y_train,
    epochs=30,
    batch_size=32,
    validation_data=([X_val[:, i] for i in range(X_val.shape[1])], y_val),
    callbacks=[lr_scheduler]
)


model.save('/Users/farazhussain/musicmodel.h5')
```
